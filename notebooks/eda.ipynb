```python
for i in cols:
    if(df[i].dtype=="int64" or df[i].dtype=="float64"):  
        num_cols.append(i)
    else:
        cat_cols.append(i)
```

```python
for col in num_cols:
    plt.figure(figsize=(12,4))
    
    # Histogram
    plt.subplot(1,2,1)
    sns.histplot(df[col], kde=True)
    plt.title(f'{col} Distribution')
    
    # Boxplot
    plt.subplot(1,2,2)
    sns.boxplot(x=df[col])
    plt.title(f'{col} Boxplot')
    
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for col in num_cols:
    print(f"--- {col} ---")
    
    # Summary statistics
    print(df.groupby('Customer Status')[col].describe())
    
    # Create a figure with 1 row, 2 columns
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))
    
    # Boxplot on the first subplot
    sns.boxplot(x='Customer Status', y=col, data=df, ax=axes[0])
    axes[0].set_title(f"{col} vs Churn (Boxplot)")
    
    # KDE plot on the second subplot
    sns.histplot(df[df['Customer Status']=='Stayed'][col], label='Not Churned',kde=True, fill=True, ax=axes[1])
    sns.histplot(df[df['Customer Status']=='Churned'][col], label='Churned', kde=True,fill=True, ax=axes[1])
    axes[1].set_title(f"{col} Distribution by Churn")
    axes[1].legend()
    
    # Adjust layout so titles and labels don't overlap
    plt.tight_layout()
    plt.show()
```

```python

for col in cat_cols:
    # Crosstab for counts
    counts = pd.crosstab(df[col], df['Customer Status'])
    
    # Churn rate per category
    churn_rate = df.groupby(col)['Customer Status'].apply(lambda x: (x=='Churned').mean() * 100)
    
    # Create side-by-side subplots
    fig, axes = plt.subplots(1, 2, figsize=(12,4))
    
    # Left: stacked bar (counts)
    counts.plot(kind='bar', stacked=True, ax=axes[0])
    axes[0].set_title(f"{col} vs Churn (Counts)")
    axes[0].set_ylabel("Number of Customers")
    
    # Right: churn rate %
    churn_rate.plot(kind='bar', color='salmon', ax=axes[1])
    axes[1].set_title(f"{col} Churn Rate (%)")
    axes[1].set_ylabel("Churn Rate (%)")
    
    plt.tight_layout()
    plt.show()
```

```python
corr_matrix = df[num_cols].corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title("Correlation Heatmap of Numerical Features")
plt.show()

threshold = 0.8
high_corr = corr_matrix.abs() > threshold

high_corr_values = np.where(high_corr & (corr_matrix.abs() < 1))
high_corr_pairs = [(num_cols[i], num_cols[j]) for i,j in zip(*high_corr_values) if i < j]

print("Highly correlated feature pairs (|corr| > 0.8):")
print(high_corr_pairs)

features_to_drop = [pair[1] for pair in high_corr_pairs]
print("\nSuggested features to drop to reduce redundancy:")
print(features_to_drop)

df_reduced = df.drop(columns=features_to_drop)
print("\nColumns after dropping highly correlated features:")
print(df_reduced.columns.tolist())
```

```python
# Target + churn rate
status = df["Customer Status"].astype(str).str.strip().str.lower()
df["is_churn"] = (status == "churned").astype(int)

print("Customer Status counts:")
print(df["Customer Status"].value_counts(dropna=False))

print("\nChurn rate (%):", round(df["is_churn"].mean() * 100, 2))
```

```python
# Cell 2: Leakage columns
leakage_cols = [c for c in ["Customer Status", "Churn Category", "Churn Reason"] if c in df.columns]
print("Leakage/label columns (exclude from features):", leakage_cols)

# Optional: confirm churn-only population for these cols
for c in ["Churn Category", "Churn Reason"]:
    if c in df.columns:
        print(f"\nNon-null rate of '{c}' by Customer Status:")
        print(df.groupby("Customer Status")[c].apply(lambda s: s.notna().mean()).sort_index())
```

```python
#Duplicates
print("Duplicate rows:", int(df.duplicated().sum()))

if "Customer ID" in df.columns:
    print("Duplicate Customer ID:", int(df["Customer ID"].duplicated().sum()))
    # Show a few duplicate IDs if any
    dup_ids = df.loc[df["Customer ID"].duplicated(keep=False), "Customer ID"]
    if not dup_ids.empty:
        print("\nExample duplicate Customer IDs:")
        print(dup_ids.value_counts().head(10))
```

```python
sample = df.sample(n=min(len(df), 2000), random_state=42)

plt.figure(figsize=(12, 5))
sns.heatmap(sample.isna(), cbar=False)
plt.title("Missingness heatmap (sampled rows)")
plt.xlabel("Columns")
plt.ylabel("Rows (sample)")
plt.tight_layout()
plt.show()
```

```python
#Numeric sanity checks
num_cols = df.select_dtypes(include=["number"]).columns.tolist()
print("Numeric columns:", num_cols)

# Check for negatives (common sanity rule for charges/tenure-like fields)
neg_counts = (df[num_cols] < 0).sum().sort_values(ascending=False)
neg_counts = neg_counts[neg_counts > 0]
print("\nNegative value counts (if any):")
print(neg_counts if not neg_counts.empty else "None")

# Check extreme values via quantiles
q = df[num_cols].quantile([0.0, 0.01, 0.05, 0.5, 0.95, 0.99, 1.0]).T
q.columns = ["min", "p01", "p05", "p50", "p95", "p99", "max"]
print("\nQuantiles (min/p01/p05/median/p95/p99/max):")
print(q.sort_values("max", ascending=False).head(20))
```

```python
# Categorical cardinality (unique counts)
cat_cols = df.select_dtypes(exclude=["number"]).columns.tolist()
card = df[cat_cols].nunique(dropna=False).sort_values(ascending=False)

print("Categorical columns unique counts (including NaN):")
print(card)

high_card = card[card > 30]
print("\nHigh-cardinality categorical columns (>30 unique):")
print(high_card if not high_card.empty else "None")
```
